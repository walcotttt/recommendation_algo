## Bivariate Exploratory Analysis

```python
# scatter plot of release_date vs. len

sns.scatterplot(data=df, x="release_date", y="len", hue="genre", alpha=0.5)
sns.color_palette("bright")
plt.show()
```

```python
# scatter plot of dating vs. romantic

sns.scatterplot(data=df, x="dating", y="romantic", hue="genre", alpha=0.7)
plt.title("Dating vs. Romantic")
plt.show()
```

```python
# scatter plot of violence vs. obscene

sns.scatterplot(data=df, x="violence", y="obscene", hue="genre", alpha=0.7)
plt.title("Violence vs. Obscene")
plt.show()
```

```python
# scatter plot of sadness vs. feelings

sns.scatterplot(data=df, x="sadness", y="feelings", hue="genre", alpha=0.7)
plt.title("Sadness vs. Feelings")
plt.show()
```

```python
# scatter plot of night/time vs. shake the audience

sns.scatterplot(data=df, x="night/time", y="shake the audience", hue="genre", alpha=0.7)
plt.title("Night vs. Shake")
plt.show()
```

```python
# scatter plot of family/gospel vs. family/spiritual

sns.scatterplot(data=df, x="family/gospel", y="family/spiritual", hue="genre", alpha=0.7)
plt.title("Gospel vs. Spiritual")
plt.show()
```

## Multivariate Exploratory Analysis

```python
# correlation heat map of selected columns
numeric_cols = ["release_date", "len", "dating", "violence", "world/life", "night/time", 
                "shake the audience", "family/gospel", "romantic", "communication", "obscene", 
                "music", "movement/places", "light/visual perceptions", "family/spiritual", "sadness", "feelings", "age"]

# correlation matrix of the numerical variables
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(18,18))
plt.title("Correlation Matrix of Numeric Variables")
sns.heatmap(corr_matrix, cmap = 'PiYG', annot=True, linewidths=0.5)
```
